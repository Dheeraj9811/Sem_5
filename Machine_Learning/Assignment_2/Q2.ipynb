```python
#making points for data x and y
x =[0,0,1,1]
y =[0,1,0,1]
```

XOR_database

```python

xor_r = [0,1,1,0]
df_xor = pd.DataFrame(list(zip(x,y,xor_r)) , columns=['X','Y','Label'])
```

```python
def PTA (p,data,flag =0):
        
        try:    
                weight, biased = p.learning(p.X_actual,p.Y_actutal,flag)
                print("Weight:",weight)
                print("Bias :", biased)
                # weights[0]*x + weight[1]*y + b = 0;

                x = np.linspace(-1,2,50)
                
                y = (-weight[0]*x-biased)/weight[1]
                
                plt.plot(x,y,'-r')
                sea.scatterplot(x='X',y='Y',data = data,hue = 'Label')
        except:
                print("Truth value of a DataFrame is ambiguous.")
        plt.axis("equal")
        plt.show()
```

```python
percep_xor = utils.Perceptron(df_xor.iloc[:,:2].values , df_xor.iloc[:,-1].values, epochs= 200)
PTA(percep_xor,0,df_xor)
```
