```python
from skimage.io import imread,imshow,imsave
import numpy as np
from matplotlib import pyplot as plt
img = imread("ruler.512_2.TIFF")
```

```python
import made_filter
save_mean_3 = made_filter.mean_filter(img,3)
save_mean_5 = made_filter.mean_filter(img,5)
```

```python
plt.figure()
plt.subplots(figsize=(15, 5))
plt.subplot(1,3,1)
plt.imshow(img,cmap=plt.get_cmap('gray'))
plt.title("original")

plt.subplot(1,3,2)
plt.imshow(save_mean_3,cmap=plt.get_cmap('gray'))
plt.title("mean filter K=3")

plt.subplot(1,3,3)
plt.imshow(save_mean_5,cmap=plt.get_cmap('gray'))
plt.title("mean filter K=5")

plt.tight_layout()
plt.show()
```
