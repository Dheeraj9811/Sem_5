```python
from skimage.io import imread,imshow,imsave
import numpy as np
from matplotlib import pyplot as plt
# import made_filter
import numpy as np
```

```python
img = imread("clock.tiff")
print(" image shape",np.shape(img))
plt.figure()
plt.subplots(figsize=(15, 5))
plt.subplot(1,3,1)
plt.imshow(img,cmap=plt.get_cmap('gray'))
plt.title("original")
```

```python
# According to question  implement nearest neighbour interpolation 
def nearest_neighbour_interpolation(img, scale):
    # get the shape of the image
    h, w = img.shape
    # get the shape of the new image
    new_h = int(h * scale)
    new_w = int(w * scale)
    # create the new image here
    
    new_img = np.zeros((new_h, new_w), dtype=np.uint8)
    # fill the new image
    for i in range(new_h):
        for j in range(new_w):
            # get the nearest neighbour
            x = int(i / scale)
            y = int(j / scale)
            # fill the new image
            new_img[i, j] = img[x, y]
    return new_img
```

```python
#acoording to question Image 4 time  so our scale = 2 
new_img = nearest_neighbour_interpolation(img , 2)
```

```python
# ploting new image \n",

print(" image shape",np.shape(new_img))
plt.figure()
plt.subplots(figsize=(15, 5))
plt.subplot(1,3,1)
plt.imshow(new_img,cmap=plt.get_cmap('gray'))
plt.title("new 4 times image ")
```
